```python
import random

try:
    # Step 1: Define lists of first and last names
    first_names = [
        "James", "Mary", "John", "Patricia", "Robert", "Jennifer", "Michael", "Linda",
        "William", "Elizabeth", "David", "Barbara", "Richard", "Susan", "Joseph", "Jessica",
        "Thomas", "Sarah", "Charles", "Karen", "Daniel", "Nancy", "Matthew", "Lisa",
        "Anthony", "Betty", "Mark", "Sandra", "Donald", "Ashley", "Steven", "Kimberly",
        "Paul", "Emily", "Andrew", "Donna", "Joshua", "Michelle", "Kenneth", "Carol"
    ]

    last_names = [
        "Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller", "Davis",
        "Rodriguez", "Martinez", "Hernandez", "Lopez", "Gonzalez", "Wilson", "Anderson",
        "Thomas", "Taylor", "Moore", "Jackson", "Martin", "Lee", "Perez", "Thompson",
        "White", "Harris", "Sanchez", "Clark", "Ramirez", "Lewis", "Robinson", "Walker",
        "Young", "Allen", "King", "Wright", "Scott", "Torres", "Nguyen", "Hill"
    ]

    # Step 2: Create a list of 400 workers dynamically
    workers = []
    genders = ['Male', 'Female']

    for i in range(1, 401):
        full_name = f"{random.choice(first_names)} {random.choice(last_names)}"
        worker = {
            'id': i,
            'name': full_name,
            'gender': random.choice(genders),
            'salary': random.randint(5000, 30000)  # Random salary between 5k and 30k
        }
        workers.append(worker)

    # Step 3: Generate payment slips for each worker
    for w in workers:
        name = w['name']
        gender = w['gender']
        salary = w['salary']

        # Conditional logic for employee level
        if 10000 < salary < 20000:
            level = "A1"
        elif 7500 < salary < 30000 and gender == "Female":
            level = "A5-F"
        else:
            level = "B2"

        # Print payment slip
        print(f"Payment Slip for {name}")
        print(f"Gender: {gender}")
        print(f"Weekly Salary: ${salary}")
        print(f"Employee Level: {level}")
        print("-" * 40)

except Exception as e:
    print("An error occurred while generating payment slips:", e)

```
